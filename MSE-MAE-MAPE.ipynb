```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,mean_absolute_percentage_error
```

```python
df = pd.read_csv('medical_cost.csv')
```

```python
df.head(3)
```

```python
df=pd.get_dummies(df, columns=['sex','smoker','region'], drop_first=True)
```

```python
df.head(3)
```

```python
y=df['charges']
x=df.drop('charges', axis=1)
```

```python
lm=LinearRegression()
model=lm.fit(x,y)
```

```python
model.predict([[20,26.5,0,1,0,0,0,1]])
```

# description behind - mse, mae, mape

```python
df_error=pd.DataFrame()
```

```python
df_error['y']=y
```

```python
df_error
```

```python
y_predict=model.predict(x)
```

```python
df_error['predict']=y_predict
```

```python
df_error['error']=y-y_predict
```

```python
df_error
```

```python
df_error['squared_error']=df_error['error']**2
```

```python
df_error
```

```python
df_error['abs_error']=np.abs(df_error['error'])
```

```python
df_error
```

```python
df_error['percent_error']=np.abs((y-y_predict)/y)
```

```python
df_error
```

```python
df_error.mean()
```
